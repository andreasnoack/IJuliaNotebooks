###High-level
A broad definition of a high-level programming language is a programming language that uses **abstractions** to hide the details of a the computer. Different chips need different machine code, but a high-level programming language hides this by generating machine code from machine independent code. First widespread high-level language is Fortran. 

Is Fortran high-level?

In original defition, yes, but now high-level often refers to languages that e.g. handles memory for the user like Python, MATLAB, R, and Julia. Hence, you don't have to explicitly allocate memory. Actually, you can't manage memory explicitly in Julia. Arrays are allocated automatrically, as in Python, R and MATLAB

Julia is high-level, but not as high as Python and R where the machine representation of numbers is made almost invisible to the user. In Julia you can choose to ignore machine representations of numbers, but often it can be useful to pay some attention to the machine representation. The reasons for this is performance (which is the reason Numpy actually makes this visible) and more details will be given in next section. The basic number types in Julia are 
- `Float64` for double precision
- `Float32` for single precision
- `Int64` for 64 bit integers
- `Int32` for 32 bit integers

but there are also smaller bits versions of floating point numbers and integers as well as unsigned integers. However, these are typically not that relevant for numerical computing. Julia defined the type `Int` to be either `Int32` or `Int64` depending on your machines register size. These days most computers are 64 bit, but Julia can in theory run on a cell phone where 32 bit is the standard.

Julia will clasify input numbers depending on how they are written in a program. Number withouth decimal points are assumed to be `Int`s and number with a decimal point will be `Float64` bit on a 64 bit system and `Float32` on a 32 bit system. There is a literal representation of `Float32`s for the cases where these are specifically wanted.

###High-performance
Julia can be almost as fast as C and Fortran because of the way the language has been designed. Because C and Fortran are considered the fastest possible languages the term high-performance seems appropriate for Julia. 

Julia is build on top of the Low Level Virtial Machine (LLVM) which is the a result of a research project at University of Illinois at Urbana-Champaign and now the backend in e.g. Apple's Xcode. LLVM can be thought of as a low-level programming language that Julia translates into. Then the LLVM code is compiled into actual machine code afterwards. LLVM also provides the backend for other project seeking to make high-level languages fast, e.g. Numba for Python and recent innovations to R. Rumors are that MATLAB is also transitioning into using LLVM for the just-in-time compiler, but I haven't confirmed that.

However, to be able to generate good LLVM code, Julia has to to able to analyse the program well. Therefore, the design of the Julia language also plays an important role. From a computer science perspective, Julia has some interesting features that enables it to fast and high-level at the same time, but I am only just starting to understand these myself. The important point to make here is that Julia is from designed to work with LLVM from the outset whereas Numba and the recent R extension are add-ons on existing languages designed with a completely different goal.

###Dynamic
In computer science, dynamic can refer to many things, but it is often used when discussing **compile time** vs **runtime** features. In Julia, dynamic mainly refers to the fact that the language is **dynamically** typed in contrast to **statically** typed. These concepts can cause long discussions for computer scientists, but can to a large extend be ignored by (economist) users of Julia. The main point is that, in Julia, you do not have to specify the type of a variable, but you are free to do so. This is in contrast to C and Fortran where types of variables have to be declared.

```julia
let # creates a local scope which avoids a global variable
    x = 1
    println(x)
end
```

```julia
let
    x::Int = 1
    println(x)
end
```

The following example shows how type annotation will cause a conversion of the value

```julia
let
    x::Int = 1.0
    println(x)
end
```

In contrast to C, conversion is only possible if no precision is lost. The following example results in an error

```julia
let
    x::Int = 1.5
    println(x)
end
```

Instead you have to explicitly truncate the floating point value

```julia
let
    x::Int = trunc(Int, 1.5)
    println(x)
end
```

## Example 2: Time Series Econometrics: ARCH
### Parametric types, Optimization, and greek letters
The model
$$x_t = \sigma_tz_t\\
\sigma_t^2 = \sigma_0 + \alpha x_{t-1}^2$$

```julia
using PyPlot
using Distributions

module ARCHModels

using Distributions

type ARCH{k,D<:Distribution{Univariate,Continuous},T<:Real}
    pars::Vector{T}
    dist::D
    function ARCH(pars::Vector{Float64}, dist::Distribution{Univariate,Continuous})
        length(pars) > 1 || throw(ArgumentError("parameter vector must have length at least two but had length: $(length(pars))"))
        return new(pars, dist)
    end
end
ARCH{T<:Real}(pars::Vector{T}, dist::Distribution{Univariate,Continuous}) = ARCH{length(pars) - 1, typeof(dist),T}(pars, dist)

import Base: rand, var

var{k,D}(model::ARCH{k,D}) = model.pars[1]/(1 - sum(model.pars[2:end].^2))

function rand{D}(model::ARCH{1,D}, T::Integer, x₀ = 0.0)
    σ₀2,α = model.pars
    σzinv = inv(var(model.dist))
    vx    = Array(Float64, T)
    vσ    = similar(vx)
    σ     = sqrt(σzinv*(σ₀2 + α*x₀*x₀))
    vx[1] = rand(model.dist)*σ
    for t = 2:T
        σ = sqrt(σzinv*(σ₀2 + α*vx[t-1]*vx[t-1]))
        vx[t] = rand(model.dist)*σ
        vσ[t] = σ
    end
    vx, vσ 
end

import Distributions: loglikelihood

function loglikelihood(θ, x, ::Type{ARCH{1,Normal}})
    σ₀2, α = θ
    T = length(x)
    σ2 = similar(x, eltype(θ), T)
    σ2[1] = σ₀2;
    ll = log(σ₀2) + x[1]*x[1]/σ₀2
    for t = 2:T
        σ2t = σ₀2 + α*x[t-1]*x[t-1]
        ll += log(σ2t) + x[t]*x[t]/σ2t
        σ2[t] = σ2t
    end
    return -0.5*ll
end

end
```

```julia
m = ARCHModels.ARCH([1.0, 1.1], Normal())
@time x, s = ARCHModels.rand(m, 1000);
```

```julia
var(m)
```

```julia
plot(x)
```

```julia
plot(s)
```

```julia
loglikelihood([1.0, 1.99], x, ARCHModels.ARCH{1,Normal})
```

```julia
linspace(0.7, 1.3, 100) |> t -> plot(t, map(s -> loglikelihood([1.0, s], x, ARCHModels.ARCH{1,Normal}), t));
```

```julia
using Optim
```

```julia
o = optimize(t -> -loglikelihood([t[1],t[2]], x, ARCHModels.ARCH{1,Normal})/length(x), [.5, 1.5], method = :l_bfgs)
```

Unrestricted optimization fails because the restriction $\alpha\geqslant 0$ is not enforced. Possible solution is to square the parameter $\alpha$.

```julia
o = optimize(t -> -loglikelihood([t[1],t[2]^2], x, ARCHModels.ARCH{1,Normal})/length(x), [.5, 1.5], method = :l_bfgs)
```

The Calculus package can be used for evaluating the numerical hession in order to get the standard errors of the estimator.

```julia
using Calculus
```

```julia
se = sqrt(diag(inv(cholfact(hessian(t -> -loglikelihood([t[1],t[2]^2], x, ARCHModels.ARCH{1,Normal}), o.minimum)))))
```

```julia
tStats = o.minimum./se
```
