#The distribution of the real roots of random polynomials
Load PyPlot and read in data from serialized objects

```julia
using PyPlot
fh = open("simG.jls")
simG = deserialize(fh)
close(fh)
fh = open("simR.jls")
simR = deserialize(fh)
close(fh)
fh = open("simU.jls")
simU = deserialize(fh)
close(fh);
```

##Redo plot of expected number of real roots for different distributions

```julia
nvals = [300:50:1000]
plot(nvals, [2/π*log(n) + 0.612738072 + 2/π/n for n in nvals])
plot(nvals, [mean([length(a) for a in b]) for b in simG])
plot(nvals, [mean([length(a) for a in b]) for b in simR])
plot(nvals, [mean([length(a) for a in b]) for b in simU])
legend(["Gaussian (analytical)", "Gaussian", "Rademacher", "Uniform"], "upper left")
title("Mean of 5000 simulations of the number of real roots of polynomials\n with idenpendent random coefficients with zero mean and\n unit variance of the specified distribution");
```

##Estimation distribution of the real roots
First with a bit rough choice of bandwidth

```julia
using KernelDensity # written by our visitor Simon Byrne
dG1000 = kde([simG[15]...][abs([simG[15]...]) .< 3])
dR1000 = kde([simR[15]...][abs([simR[15]...]) .< 3])
dU1000 = kde([simU[15]...][abs([simU[15]...]) .< 3]);
xvals = linspace(-3, 3, 100)
xlim(-2.5,2.5)
plot(xvals, [1/π*sqrt(1/(x^2 - 1)^2) - (1000 + 1)^2*x^(2*1000)/(x^(2*1000 + 2) - 1)^2 for x in xvals])
plot(dG1000.x, dG1000.density*mean([length(n) for n in simG[15]])) # I multiply with the estimated mean of the number of real roots
plot(dR1000.x, dR1000.density*mean([length(n) for n in simR[15]])) # I multiply with the estimated mean of the number of real roots
plot(dU1000.x, dU1000.density*mean([length(n) for n in simU[15]])) # I multiply with the estimated mean of the number of real roots
legend(["Gaussian (analytical)", "Gaussian", "Rademacher", "Uniform"], "upper left")
title("Kernel density estimates of the distribution of the real roots\n of polynomials with independent random coefficients\n with the specified distribution. Silverman's bandwith has been used.");
```

Next plot has a smaller bandwidth that better allows the estimated density to follow the theoretical distribution around the poles

```julia
dG1000 = kde([simG[15]...][abs([simG[15]...]) .< 3], bandwidth = 0.01)
dR1000 = kde([simR[15]...][abs([simR[15]...]) .< 3], bandwidth = 0.01)
dU1000 = kde([simU[15]...][abs([simU[15]...]) .< 3], bandwidth = 0.01);
xlim(-2.5, 2.5)
ylim(0, 3)
plot(xvals, [1/π*sqrt(1/(x^2 - 1)^2) - (1000 + 1)^2*x^(2*1000)/(x^(2*1000 + 2) - 1)^2 for x in xvals])
plot(dG1000.x, dG1000.density*mean([length(n) for n in simG[15]])) # I multiply with the estimated mean of the number of real roots
plot(dR1000.x, dR1000.density*mean([length(n) for n in simR[15]])) # I multiply with the estimated mean of the number of real roots
plot(dU1000.x, dU1000.density*mean([length(n) for n in simU[15]])) # I multiply with the estimated mean of the number of real roots
legend(["Gaussian (analytical)", "Gaussian", "Rademacher", "Uniform"], "upper left")
title("Kernel density estimates of the distribution of the real roots\n of polynomials with independent random coefficients\n with the specified distribution. Bandwidth is 0.01");
```

```julia
using GLM
using DataFrames
df = DataFrame(En = [Float64[mean([length(a) for a in b]) for b in simG], 
                Float64[mean([length(a) for a in b]) for b in simR], 
                Float64[mean([length(a) for a in b]) for b in simU]],
logn = [fill(log(300:50:1000), 3)...],
dist = pool([fill(:Gauß, 15), fill(:Rademacher, 15), fill(:Uniform, 15)]))
```

```julia
fit1 = lm(En~logn*dist - 1, df)
```

```julia
fit2 = lm(En~logn + dist, df)
```

```julia
r1 = residuals(fit1)
r2 = residuals(fit2)
ll1 = -length(r1)/2*log(dot(r1, r1)/length(r1))
ll2 = -length(r2)/2*log(dot(r2, r2)/length(r2))
lr = 2(ll1 - ll2)
@printf("Test: %f, p-value: %f\n", lr, ccdf(Chisq(2), lr))
```
